```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# To change comma to decimal in csv file, use decimal=","
df= pd.read_csv(file_path, decimal=",")

df.head()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
#We use the .describe() method. This tells Python to run descriptive statistics on every numerical column in the DataFrame.

df.describe()
```

```python
# to find max date and min date
max_date = df['date'].max()
print('The maximunm date is' + " " +str(max_date))
min_date= df['date'].min()
print('The minimum date is' + " "+ str(min_date))
```

```python
#create filter for june 1st
df_june = df[(df['date'] > "2017-05-31 23:59:59") &(df['date'] < 
                                                    "2017-06-02")].reset_index(drop=True)

# Preview the new dataframe
df_june.head()
```

```python
# 1. Create a list of the columns we want
important_cols = ['date', '% Iron Concentrate', '% Silica Concentrate', 'Ore Pulp pH', 'Flotation Column 05 Level']

# 2. Apply that list to our dataframe
df_june_important = df_june[important_cols]

# 3. View the final result
df_june_important.head()
```

```python
#"How do these variables relate to each other? Does the pH level affect the Iron Concentrate?"
sns.pairplot(df_june_important)
```

```python
# 1. Do the Math (Filter to numeric columns only)
correlation_table = df_june_important.corr(numeric_only=True)

# 2. Draw the Heatmap
sns.heatmap(correlation_table, annot=True, cmap='coolwarm')
```

```python
#"Show me exactly how the Iron Concentrate changed hour-by-hour on June 1st."
sns.lineplot(x='date', y='% Iron Concentrate', data=df_june)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6)) # Increase the chart size (width=12, height=6)

sns.lineplot(x='date', y='% Iron Concentrate', data=df_june)

plt.title('% Iron Concentrate Over Time')
plt.xlabel('Date')
plt.ylabel('% Iron Concentrate')

plt.tight_layout() # Prevent labels from overlapping
plt.show()
```

```python
# List of columns we want to plot
columns_to_plot = ['% Iron Concentrate', '% Silica Concentrate', 'Ore Pulp pH', 'Flotation Column 05 Level']

# Loop through the list
for col in columns_to_plot:
    plt.figure(figsize=(12, 6))
    sns.lineplot(x='date', y=col, data=df_june)
    plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
# Histogram with KDE (Kernel Density Estimate)
# silica concetrate distribution
plt.figure(figsize=(10,5))
sns.histplot(df['% Silica Concentrate'], bins=30,color= "Green",  kde=True)
plt.title('% Silica Concentrate Distribution')
plt.xlabel('Silica Concentrate')
plt.ylabel('Frequency')
plt.show()
```
